```python
# seperating X and y
X = df.loc[:,['Overall Life','Male Life','Female Life']].values
y = df.loc[:,['Continent']].values
```

```python
X
```

# Feature scaling

->Whenever we are dealing with KNN algorithm, we should do the feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
y=y.reshape(-1)
```

```python
y.shape
```

```python
#Splitting the data into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0, test_size = 0.30)
```

```python
#importing classification metrics
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

# Selecting K value.

```python
acc_list = []
err_list = []
for i in range(1,15):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test,y_pred)
#     print("For k = {}, accuracy = {}".format(i,acc))
    acc_list.append(acc)
    err_list.append(1-acc)
```

```python
plt.plot(list(range(1,15)),err_list,c= 'r')
plt.title('Error rate v/s K')
plt.xlabel('K')
plt.ylabel('Error rate')
plt.show()
```

# Model building

```python
model = KNeighborsClassifier(n_neighbors=9)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

# Model Evaluation

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
